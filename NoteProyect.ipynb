```python
# CONVERSOR DE MONEDA 3000 :)
print("CONVERSOR DE MONEDA ACTUALIZABLE")
print("=" * 35)

print("\nPor favor, ingresa las tasas de cambio ACTUALES:")
usd_cup = float(input("¿A cuántos CUP está el DÓLAR (USD)?: "))
eur_cup = float(input("¿A cuántos CUP está el EURO (EUR)?: "))
mxn_cup = float(input("¿A cuántos CUP está el PESO MEXICANO (MXN)?: "))

print("\n" + "=" * 35)

print("\n¿Qué conversión necesitas?")
print("1. USD → CUP")
print("2. EUR → CUP")
print("3. MXN → CUP")
print("4. CUP → USD")
print("5. CUP → EUR")
print("6. CUP → MXN")

opcion = input("\nElige una opción (1-6): ")

# Realizar la conversión según la opción
if opcion == "1":
    cantidad = float(input("\n¿Cuántos DÓLARES (USD) quieres convertir?: "))
    resultado = cantidad * usd_cup
    print(f"{cantidad} USD = {resultado} CUP")

elif opcion == "2":
    cantidad = float(input("\n¿Cuántos EUROS (EUR) quieres convertir?: "))
    resultado = cantidad * eur_cup
    print(f"{cantidad} EUR = {resultado} CUP")

elif opcion == "3":
    cantidad = float(input("\n¿Cuántos peso MEXICANOS(MXN)quieres convertir?:"))
    resultado = cantidad * mxn_cup
    print(f"{cantidad} MXN = {resultado} CUP")

elif opcion == "4":
    cantidad = float(input("\n¿Cuántos CUP quieres convertir?: "))
    resultado = cantidad / usd_cup
    print(f"{cantidad} CUP = {resultado:.2f} USD")

elif opcion == "5":
    cantidad = float(input("\n¿Cuántos CUP quieres convertir?: "))
    resultado = cantidad / eur_cup
    print(f"{cantidad} CUP = {resultado:.2f} EUR")

elif opcion == "6":
    cantidad = float(input("\n¿Cuántos CUP quieres convertir?: "))
    resultado = cantidad / mxn_cup
    print(f"{cantidad} CUP = {resultado:.2f} MXN")

else:
    print("Opción no válida")
```
